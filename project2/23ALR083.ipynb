```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = {
    "Hours_Studied": [1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0,
                      5.5, 6.0, 6.5, 7.0, 7.5, 8.0],
    "Marks": [35, 40, 48, 55, 58, 62, 68, 75,
              78, 82, 85, 88, 92, 95]
}

df = pd.DataFrame(data)
df.to_csv("student_scores.csv", index=False)
print("Dataset saved as student_scores.csv")
```

```python
df = pd.read_csv("student_scores.csv")
print(df)
```

```python
X = df[['Hours_Studied']]
y = df['Marks']
```

```python
model = LinearRegression()
model.fit(X, y)

print("Model trained successfully!")
```

```python
hours = float(input("Enter hours studied: "))
predicted_marks = model.predict([[hours]])
print(f"Predicted Marks: {predicted_marks[0]:.2f}")
```

```python
plt.scatter(X, y, label="Actual Data")
plt.plot(X, model.predict(X), label="Prediction Line")
plt.xlabel("Hours Studied")
plt.ylabel("Marks")
plt.title("Student Score Prediction")
plt.legend()
plt.show()
```
